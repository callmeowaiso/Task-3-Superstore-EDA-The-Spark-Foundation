```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'matches' with a 'season' and 'winner' column

winning_team = {}  # Create an empty dictionary to store the winning team for each season

for season in sorted(matches['season'].unique()):
    season_df = matches[matches['season'] == season]
    last_match_of_season = season_df.iloc[-1]  # Get the last match of the season
    winning_team[season] = last_match_of_season['winner']

# Convert the dictionary to a DataFrame for visualization (optional)
winning_team_df = pd.DataFrame(list(winning_team.items()), columns=['Season', 'Winner'])

# Create a bar plot to visualize the winning team of each season
plt.figure(figsize=(12, 6))
winning_team_df['Winner'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Winning Team of Each IPL Season", fontsize=16, fontweight='bold')
plt.xlabel("Team", fontsize=12)
plt.ylabel("Number of Seasons Won", fontsize=12)
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.tight_layout()
plt.show()
```

```python


# Assuming you have a DataFrame named 'deliveries' with columns 'bowler' and 'dismissal_kind'

# Filter for dismissals that indicate a wicket taken (e.g., 'bowled', 'caught', 'lbw', 'caught and bowled', etc.)
dismissals_wickets = deliveries[deliveries['dismissal_kind'].isin(['bowled', 'caught', 'lbw', 'caught and bowled', 'stumped', 'hit wicket'])]

# Calculate the total wickets taken by each bowler
bowler_wickets = dismissals_wickets.groupby('bowler').size().reset_index(name='wickets')

# Get the top N bowlers with the most wickets (change N as needed)
top_bowlers = bowler_wickets.nlargest(10, 'wickets')  # Display the top 10 bowlers, adjust as desired

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_bowlers['bowler'], top_bowlers['wickets'], color='pink')
plt.title("Top Bowlers by Total Wickets Taken", fontsize=16, fontweight='bold')
plt.xlabel("Bowler", fontsize=12)
plt.ylabel("Total Wickets Taken", fontsize=12)
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'deliveries' with columns 'bowler', 'total_runs', and 'player_dismissed'

# Calculate the total runs conceded by each bowler
bowler_runs_conceded = deliveries.groupby('bowler')['total_runs'].sum()

# Calculate the total wickets taken by each bowler
bowler_wickets_taken = deliveries[deliveries['player_dismissed'].notna()].groupby('bowler').size()

# Calculate the total balls bowled by each bowler
bowler_balls_bowled = deliveries.groupby('bowler').size()

# Calculate the economy rate for each bowler (runs conceded per over)
economy_rate = (bowler_runs_conceded / (bowler_balls_bowled / 6)).round(2)

# Calculate the bowling average for each bowler (runs conceded per wicket)
bowling_average = (bowler_runs_conceded / bowler_wickets_taken).round(2)

# Combine economy rate and bowling average to calculate a consistency score (lower is better)
consistency_score = (economy_rate + bowling_average).sort_values()

# Get the top N most consistent bowlers
top_bowlers = consistency_score.head(10)  # Display the top 10 consistent bowlers, adjust as needed

# Create a bar plot to visualize the most consistent bowlers
plt.figure(figsize=(12, 6))
top_bowlers.plot(kind='bar', color='purple')
plt.title("Most Consistent Bowlers", fontsize=16, fontweight='bold')
plt.xlabel("Bowler", fontsize=12)
plt.ylabel("Consistency Score", fontsize=12)
plt.xticks(rotation=45, ha="right")

# Show the plot
plt.tight_layout()
plt.show()
```
